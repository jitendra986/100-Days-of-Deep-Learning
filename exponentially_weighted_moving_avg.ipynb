```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import drive

drive.mount('/content/drive')

#df = pd.read_csv('placement.csv')
df = pd.read_csv('/content/drive/My Drive/Colab_Notebooks/DailyDelhiClimateTest.csv')
```

```python
df = df[['date','meantemp']]
df
```

```python
plt.scatter(df['date'],df['meantemp'],color='red')
```

```python
#to calculate exponentially weighted moving average
#if alpha=0.9 it means beta=0.1
x1=df['meantemp'].ewm(alpha=0.1).mean()
x1
```

```python
df['ewma']=x1
df
```

```python
plt.scatter(df['date'],df['meantemp'],color='red')
plt.plot(df['date'],df['ewma'],color='black',label='alpha=0.9')
plt.legend()
plt.show()
```
